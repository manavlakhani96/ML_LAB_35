```python
## Program-1 : To practice python numpy functions.
import numpy as np
```

```python
# Creating a 1D array
array_1d = np.array([1, 2, 3, 4, 5])    
print(f"Array from list: {array_1d}")
```

```python
#Creating a 2D array
array_2d = np.array([[1, 2, 3], [4, 5, 6]])
print(f"2D Array:\n {array_2d}")
```

```python
#using arange, linescpace, zeros, ones, eye
arr_arrange = np.arange(0, 10, 2)
print(f"Arange Array : {arr_arrange}")
```

```python
arr_linespace = np.linspace(0, 1, 5)
print(f"Linespace Array : {arr_linespace}")
```

```python
arr_zeors = np.zeros((2, 3))
print(f"Zeros Array : \n{arr_zeors}")
```

```python
arr_ones = np.ones((3, 2), dtype=int)
print(f"Ones Array : \n{arr_ones}")
```

```python
arr_eye = np.eye(3, dtype=int)
print(f"Identity Matrix : \n{arr_eye}")
```

```python
# Array Attributes
print(f"Shape of 2D Array: {array_2d.shape}")
print(f"Size of 2D Array : {array_2d.size}")
print(f"Dimensions of 2D Array : {array_2d.ndim}")
print(f"Data type of 2D Array : {array_2d.dtype}")
print(f"Item size of 2D Array : {array_2d.itemsize} bytes")
```

```python
#Indexing and Slicing
arr = np.array([10, 20, 30, 40, 50])
print(f"Original Array: {arr}")
print(f"Element at index 2: {arr[2]}")
print(F"Last Element: {arr[-1]}")
print(f"Sliced Array (1 to 4): {arr[1:4]}")
```

```python
# 2D Array Slicing
arr2d = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])
print(f"Original 2D Array:\n{arr2d}")
print(f"First Row: {arr2d[0, :]}")
print(f"Second column: {arr2d[:, 1]}")
print(f"Sunmatrix : \n{arr2d[0:2, 1:3]}")
```

```python
# Basic Operations
a = np.array([1,2,3])
b = np.array([4,5,6])

print("a + b:", a + b)
print("a - b:", a - b)
print("a * b:", a * b)
print("a / b:", a / b)
print("a ** 2:", a**2)
```

```python
#Broadcasting
c = np.array([[1,2,3], [4,5,6]])
d = np.array([1,1,1])
print("Broadcasting result:\n", c + d)
```

```python
# Universal Functions
arr = np.array([1,4,9, 16])
print("sqrt:", np.sqrt(arr))
print("exp:", np.exp(arr))
print("log:", np.log(arr))
print("sin:", np.sin(arr))
print("cos:", np.cos(arr))
```

```python
# Aggreation Functions
arr = np.array([[1,2,3],[4,5,6]])
print("Sum:", np.sum(arr))
print("Sum axis 0:", np.sum(arr, axis=0))
print("Sum axis 1:", np.sum(arr, axis=1))
print("Mean:", np.mean(arr))
print("Max:", np.max(arr))
print("Min:", np.min(arr))
print("Argmax:", np.argmax(arr))
print("Argmin:", np.argmin(arr))
print("Standard deviation:", np.std(arr))
print("Variance:", np.var(arr))
```

```python
# Reshaping and Flattenting
arr = np.arange(12)
print("Original array:", arr)
arr_reshaped = arr.reshape(3,4)
print("Reshaped 3x4 array:\n", arr_reshaped)
arr_flat = arr_reshaped.flatten()
print("Flattened array:", arr_flat)
arr_transpose = arr_reshaped.T
print("Transpose:\n", arr_transpose)
```

```python
# Stacking and Splitting
a = np.array([1,2,3])
b = np.array([4,5,6])
print("Horizontal stack:", np.hstack((a,b)))
print("Vertical stack:\n", np.vstack((a,b)))

arr = np.arange(9)
print("Original array:", arr)
print("Split into 3 parts:", np.split(arr, 3))
```

```python
#  Copy V/S View
arr = np.array([1,2,3,4])
arr_view = arr.view()
arr_copy = arr.copy()
arr_view[0] = 100
arr_copy[1] = 200
print("Original after view change:", arr)
print("Original after copy change:", arr)
```

```python
# Boolean Indexing & Fancy Indexing
arr = np.array([10, 20, 30, 40, 50])
print(f"Elements > 25 : {arr[arr > 25]}")
indices = [0,2,4]
print(f"Fancy indexing : {arr[indices]}")
```

```python
# Random
rand_arr = np.random.rand(3,3) # uniform [0,1)
print("Random array:\n", rand_arr)

rand_int = np.random.randint(0,10,5)
print("Random integers:", rand_int)

normal_arr = np.random.randn(3,3) # standard normal
print("Normal distribution array:\n", normal_arr)
```

```python
# Linear Algebra Functions
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print("Matrix multiplication:\n", np.dot(A,B))
print("Matrix determinant:", np.linalg.det(A))
print("Matrix inverse:\n", np.linalg.inv(A))
print("Eigenvalues and Eigenvectors:\n", np.linalg.eig(A))
```
