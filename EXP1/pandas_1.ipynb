```python
# Program-2 to practice python pandas functions.
import pandas as pd
import numpy as np
```

```python
# Series from list
s1 = pd.Series([10, 20, 30, 40])
print(f"Series from list:\n{s1}")

# Series with custom index
s2 = pd.Series([1, 2, 3], index=['a','b','c'])
print(f"\nSeries with custom index:\n{s2}", )

# Series from dictionary
s3 = pd.Series({'x': 100, 'y': 200})
print(f"\nSeries from dictionary:\n{s3}")
```

```python
# Dataframe from dictionary
df1 = pd.DataFrame({
'Name': ['Alice','Bob','Charlie'],
'Age': [25,30,35],
'City': ['NY','LA','Chicago']
})
print("DataFrame from dictionary:\n", df1)
```

```python
# From NumPy array
arr = np.arange(9).reshape(3,3)
df2 = pd.DataFrame(arr, columns=['A','B','C'])
print(f"\nDataFrame from NumPy array:\n{df2}")
```

```python
# From list of lists
df3 = pd.DataFrame([[1,2],[3,4],[5,6]], columns=['X','Y'])
print(f"\nDataFrame from list of lists:\n{df3}")
```

```python
# Data Viewing 
print("Head of df1:\n", df1.head(2))
print("Tail of df1:\n", df1.tail(2))
print("Columns of df1:", df1.columns)
print("Index of df1:", df1.index)
print("Shape of df1:", df1.shape)
print("Info of df1:")
print(df1.info())
print("Describe numeric data in df1:\n", df1.describe())
```

```python
# Selecting
print(f"Select 'Name' column:\n{df1['Name']}")
print("\n")
print(f"Select multiple columns:\n{df1[['Name','City']]}")
print("\n")
print(f"Select row by loc:\n{df1.loc[1]}")
print("\n")
print(f"Select row by iloc:\n{df1.iloc[0]}")
print("\n")
print(f"Select specific element:\n{df1.loc[1,'City']}")
```

```python
# Filtering
print(f"Rows where Age > 28:\n{df1[df1['Age']>28]}")
print("\n")
print(f"Rows where City is 'NY':\n{df1[df1['City']=='NY']}")
```

```python
# Adding & Modifying Columns
df1['Salary'] = [50000, 60000, 70000]
print(f"After adding Salary:\n{df1}")
print("\n")
df1['AgePlus5'] = df1['Age'] + 5
print(f"After modifying/adding AgePlus5:\n{df1}")
```

```python
# Dropping Columns/Rows
df_dropped = df1.drop('AgePlus5', axis=1)
print(f"After dropping column AgePlus5:\n{df_dropped}")
print(*"\n")
df_dropped_row = df1.drop(0, axis=0)
print(f"After dropping row 0:\n{df_dropped_row}")
```

```python
# Handling Missing Data
df_missing = pd.DataFrame({
'A':[1,np.nan,3],
'B':[4,5,np.nan]
})
print(f"DataFrame with missing values:\n{df_missing}")
print("\n")
print(f"Drop rows with NaN:\n{df_missing.dropna()}")
print("\n")
print(f"Fill NaN with 0:\n{df_missing.fillna(0)}")
print("\n")
print(f"Check for NaN:\n{df_missing.isna()}")
```

```python
# Aggreation
print(f"Sum of numeric columns:\n{df1[['Age','Salary']].sum()}")
print("\n")
print(f"Mean of numeric columns:\n{df1[['Age','Salary']].mean()}")
print("\n")
print(f"Max of numeric columns:\n{df1[['Age','Salary']].max()}")
print("\n")
print(f"Min of numeric columns:\n{df1[['Age','Salary']].min()}")
print("\n")
print(f"Describe:\n{df1[['Age','Salary']].describe()}")
```

```python
# Grouping & Aggreation
df_group = pd.DataFrame({
'Department':['HR','IT','HR','IT','IT'],
'Employee':['Alice','Bob','Charlie','David','Eve'],
'Salary':[50000,60000,55000,65000,70000]
})
print(f"Original DataFrame:\n{df_group}")
grouped = df_group.groupby('Department')['Salary'].mean()
print(f"Average salary by Department:\n{grouped}")
```

```python
# Sorting
print(f"Sort by Age ascending:\n{df1.sort_values('Age')}")
print("\n")
print(f"Sort by Salary descending:\n{df1.sort_values('Salary', ascending=False)}")
```

```python
# Merging & Joining
df_a = pd.DataFrame({'ID':[1,2,3],'Name':['A','B','C']})
df_b = pd.DataFrame({'ID':[2,3,4],'Score':[90,80,70]})
merged_df = pd.merge(df_a, df_b, on='ID', how='outer')
print(f"Merged DataFrame (outer join):\n{merged_df}")
print("\n")
concat_df = pd.concat([df_a, df_b], ignore_index=True)
print(f"Concatenated DataFrame:\n{concat_df}")
```

```python
# File I/O
# Save to CSV
df1.to_csv('df1.csv', index=False)
print("DataFrame saved to 'df1.csv'")
# Read from CSV
df_loaded = pd.read_csv('df1.csv')
print("Loaded DataFrame:\n", df_loaded)

print("\nAll major Pandas functions executed successfully!")
```
